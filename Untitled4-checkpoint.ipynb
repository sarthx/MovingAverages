```python
import talib
import yfinance as yf
import matplotlib.pyplot as plt

Tesla = yf.download("TSLA", start="2020-04-01", end="2021-04-15")
plt.plot(Tesla['Close'], color='blue')
plt.title("Daily Close Price")
plt.show()
```

```python
Tesla['SMA_50'] = talib.SMA(Tesla['Close'], timeperiod=50)

plt.plot(Tesla['Close'], color='blue', label='Daily Close Price')
plt.plot(Tesla['SMA_50'], color='red', label='SMA 50')
plt.legend()
plt.title('Simple Moving Averages')
plt.show()
```

```python
Tesla['EMA_50'] = talib.EMA(Tesla['Close'], timeperiod=50)

plt.plot(Tesla['Close'], color='blue', label='Daily Close Price')
plt.plot(Tesla['EMA_50'], color='green', label='EMA 50')
plt.legend()
plt.title('Exponential Moving Averages')
plt.show()
```
